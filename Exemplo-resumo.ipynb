```python
import pandas as pd
import matplotlib.pyplot as plt

# Dados de exemplo
data = {
    'Data': ['2023-01-01', '2023-02-01', '2023-03-01', '2023-04-01', '2023-05-01', '2023-06-01', '2023-07-01', '2023-08-01', '2023-09-01', '2023-10-01', '2023-11-01', '2023-12-01'],
    'Vendas': [1200, 1500, 1350, 1800, 2000, 2400, 2800, 2650, 2200, 1800, 1500, 1300]
}

# Criar DataFrame
df = pd.DataFrame(data)
df['Data'] = pd.to_datetime(df['Data'])
df = df.set_index('Data')

# Calcular a mdia mvel dos dados (Tendncia)
window = 3  # Nmero de perodos para calcular a mdia mvel
df['Tendencia'] = df['Vendas'].rolling(window).mean()

# Calcular a mdia acumulada dos dados
df['MediaAcumulada'] = df['Vendas'].expanding().mean()

# Calcular o desvio padro acumulado dos dados
df['DesvioPadraoAcumulado'] = df['Vendas'].expanding().std()

# Plotar o grfico
plt.figure(figsize=(10, 6))
ax = plt.gca()

# Plotar as vendas mensais
ax.plot(df.index, df['Vendas'], marker='o', color='blue', label='Vendas')

# Plotar a tendncia (mdia mvel)
ax.plot(df.index, df['Tendencia'], linestyle='--', color='green', label='Tendncia')

# Plotar a mdia acumulada
ax.plot(df.index, df['MediaAcumulada'], linestyle='-', color='orange', label='Mdia Acumulada')

# Plotar o desvio padro acumulado
ax.plot(df.index, df['DesvioPadraoAcumulado'], linestyle='-', color='red', label='Desvio Padro Acumulado')

plt.xlabel('Data')
plt.ylabel('Vendas')
plt.title('Sries Temporais de Vendas')
plt.legend()
plt.xticks(rotation=45)
plt.show()
```
